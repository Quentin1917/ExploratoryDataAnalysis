### Rpartition par race

```python
# Vrification des valeurs nuls
race_nul = dataset['race'].isnull().sum()
print("Nombre de valeurs manquantes dans la colonne 'race':", race_nul)
```

```python
# retire les valeurs nuls du dataset
race = dataset.dropna(subset=['race'])
race.value_counts
```

```python
# Etude du dataset sur les race
counts_field =race['race'].value_counts()
counts_field.value_counts
```

```python
# Transformation des valeurs numriques en texte pour la variable 'race'
dataset['race'] = dataset['race'].apply(lambda x: 
                                       "Black/African American" if x == 1
                                       else "European/Caucasian-American" if x == 2
                                       else "Latino/Hispanic American" if x == 3
                                       else "Asian/Pacific Islander/Asian-American" if x == 4
                                       else "Native American" if x == 5 
                                       else "Other" if x == 6                                                            
                                       else "NAN")

# Affichage des valeurs uniques pour la variable 'race'
print(dataset['race'].value_counts())
```

```python
# Cration d'un tableau des races par ges

# Cration un DataFrame avec les races et l'ge
df = pd.DataFrame({'race': dataset['race'], 'age': dataset['age']})

# Crer une table croise des races et de l'ge
tableau_races_age = pd.crosstab(index=df['race'], columns='count', values=df['age'], aggfunc='mean').reset_index()

# Trie des donnes par ge
tableau_races_age = tableau_races_age.sort_values(by='count', ascending=False)

# Visualisation dans un tableau des races / ge
fig, ax = plt.subplots(figsize=(10, 5))
ax.barh(tableau_races_age['race'], tableau_races_age['count'], color='maroon')

ax.set_xlabel("ge moyen")
ax.set_ylabel("Race")
ax.set_title("ge moyen par race")
plt.show()
```

```python
# Visualisation dans un tableau des tudes / ge
age = dataset["age"]

fig = plt.figure(figsize=(10, 5))
 
# creating the bar plot
plt.barh(field_cd, age, color='maroon') # plt.barh(y, x, color='maroon')
 
plt.xlabel("Age")
plt.ylabel("Etudes")
plt.title("Rpartition des tiants par tudes / ges")
plt.show()
```

Nombre de match

```python
# Vrification des valeurs nuls
match_nul = dataset['match'].isnull().sum()
print("Nombre de valeurs manquantes dans la colonne 'field_cd':", match_nul)
```

```python
# Transformation des valeurs numriques en texte pour la variable 'race'
dataset['match'] = dataset['match'].apply(lambda x: 
                                       "no match" if x == 0
                                       else "match" if x == 1                                                                                                  
                                       else "NAN")

# Affichage des valeurs uniques pour la variable 'race'
print(dataset['match'].value_counts())
```

```python
# Pie_chart sur les matchs

# Calculer les pourcentages
match_counts = dataset['match'].value_counts(normalize=True) * 100

# Crer le graphique interactif
fig = px.pie(match_counts, 
             values=match_counts.values, 
             names=match_counts.index,
             title='Pourcentage de matchs vs. pas de match')

# Afficher le graphique interactif
fig.show()
```

```python
# Compter les occurrences de matchs par genre
match_gender_counts = dataset.groupby(['match', 'gender']).size().unstack()

# Crer un graphique en barres empiles
match_gender_counts.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'pink'])

# Ajouter des tiquettes et un titre
plt.xlabel('Match')
plt.ylabel('Nombre de personnes')
plt.title('Comparaison des matchs en fonction du genre')
plt.xticks(rotation=0)  # Rotation des tiquettes de l'axe x

# Afficher la lgende
plt.legend(title='Genre', loc='upper right')

# Afficher le graphique
plt.show()
```
