```python
# Importer des bibliothèques utiles (pandas, seaborn, matplotlib, numpy)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Donnez la liste des variables présentes dans ce jeu de données ainsi que leur nature.

df=pd.read_csv("src/house_pricing.csv")
df
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# Lors de la création d'un nuage de points, examinez le comportement de la colonne LotArea par rapport à la colonne SalePrices .

sns.relplot(x='SalePrice', y='LotArea',data=df)
```

```python
# On peut en déduire qu'un terrain de 30.000 pieds carrés ou moins coute moins de 500.000 pour une majorité.
```

```python
# Affinez votre visualisation !

sns.relplot(x="LotArea", y="SalePrice", data=df.loc[(df['LotArea'] < 20000) & (df['SalePrice'] < 500000), :])
```

```python
# Lors de la création d'un nuage de points, examinez la relation entre LotFrontage et LotArea .

sns.relplot(x='LotFrontage', y='LotArea',data=df)
```

```python
# De la même manière, affinez votre visualisation

sns.relplot(x="LotFrontage", y="LotArea", data=df.loc[(df['LotArea'] < 100000) & (df['LotFrontage'] < 200), :])
```

### Exercice 2

```python
# Importez l'ensemble de données nommé house_pricing.csv. Découpez l'ensemble de données pour ne conserver que les 15 dernières colonnes.

cols=df.iloc[ 0: ,-15:].head(15)
df=cols
df
```

```python
df.describe(include='all')
```

```python
correlation=df.corr('pearson',numeric_only = True)
correlation
```

```python
# Créer une heatmap avec différentes valeurs de corrélations

sns.heatmap(data=correlation,annot=True, vmin=0)
```
